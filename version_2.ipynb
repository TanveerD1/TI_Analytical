#### 2. Dataset: Food_Prices_Dimension_Table.csv
This dataset provides a categorization of food products with hierarchical groupings, likely for pricing or nutritional analysis.

- **EFPG_code** (Numerical, Discrete): Unique code for each food product group.
  - Distinct Values: 10000, 10025, 10050, ..., 78000 (specific codes for each food group, ranging from 10000 to 78000)
- **EFPG_name** (Categorical, Nominal): Name of the food product group.
  - Distinct Values: Whole-grain breads, Whole-grain rice and pasta, ..., Infant formula (e.g., "Potatoes, fresh", "Fish and seafood, frozen")
- **Tier 1 group** (Categorical, Nominal): Broad category of the food product.
  - Distinct Values: Grains, Vegetables, Fruit, Dairy, Meat and Protein Foods, Prepared meals, sides, and salads, Other foods
- **Tier 2 group** (Categorical, Nominal): Subcategory within the Tier 1 group.
  - Distinct Values: Whole-grain breads, cereal, rice, pasta, and flours; Non-whole-grain breads, cereal, rice, pasta, and flours; Potatoes; ..., Infant formula (e.g., "Dark green", "Beverages")

#### 3. Dataset: Income-Expenditure.csv
This dataset contains financial and demographic information for 50 households, focusing on income, expenses, and family details.

- **Mthly_HH_Income** (Numerical, Continuous): Monthly household income in currency units.
  - Range: 5000 to 100000
- **Mthly_HH_Expense** (Numerical, Continuous): Monthly household expenses in currency units.
  - Range: 2000 to 50000
- **No_of_Fly_Members** (Numerical, Discrete): Number of family members in the household.
  - Distinct Values: 1, 2, 3, 4, 5, 6, 7
- **Emi_or_Rent_Amt** (Numerical, Continuous): Monthly EMI or rent amount in currency units.
  - Range: 0 to 35000
- **Annual_HH_Income** (Numerical, Continuous): Annual household income in currency units.
  - Range: 64200 to 1404000
- **Highest_Qualified_Member** (Categorical, Ordinal): Highest educational qualification of a household member.
  - Distinct Values: Illiterate, Under-Graduate, Graduate, Post-Graduate, Professional
- **No_of_Earning_Members** (Numerical, Discrete): Number of earning members in the household.
  - Distinct Values: 1, 2, 3, 4

# 1. Income & Food Affordability - How does household income relate to affordability of different food categories?
- Use the Inncome-Expenditure dataset to analyze household inncome and overall expenses


```python

print("Column Names:", income.columns)
income.columns = income.columns.str.strip()


if 'Highest_Qualified_Member' in income.columns:
    print("Column exists: Highest_Qualified_Member")
else:
    print("Column 'Highest_Qualified_Member' not found!")
    raise KeyError("Please ensure the column name is correct.")

income['Mthly_HH_Income'] = pd.to_numeric(income['Mthly_HH_Income'], errors='coerce')
income['Mthly_HH_Expense'] = pd.to_numeric(income['Mthly_HH_Expense'], errors='coerce')
income['No_of_Fly_Members'] = pd.to_numeric(income['No_of_Fly_Members'], errors='coerce')

income['Highest_Qualified_Member'] = income['Highest_Qualified_Member'].fillna('Unknown')

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=income,
    x='Mthly_HH_Income',
    y='Mthly_HH_Expense',
    hue='Highest_Qualified_Member',
    palette='Set1',
    size='No_of_Fly_Members',
    sizes=(50, 500),
    alpha=0.7
)

plt.title('Comparison of Household Income and Expenses')
plt.xlabel('Monthly Household Income')
plt.ylabel('Monthly Household Expenses')
plt.grid(True)
plt.legend(title='Highest Qualified Member', bbox_to_anchor=(1.05, 1), loc='upper left', ncol=1)
plt.tight_layout()
plt.show()
```

## Analysis
- Most households have incomes below 60,000 and expenses below 30,000.
- A dense cluster of households is observed at 20,000–40,000 income and 10,000–20,000 expenses.
- Graduates seem to make up the largest dempgraphic, with a significant number of households having incomes below 40000 and expenses below 20,000.
- Undergraduate and Illiterate households are mostly at lower incomes (0–40,000) and expenses (0–20,000).
- Professional/Post-Graduate households will likely maintain higher incomes/expenses; larger families may face financial pressure; lower-income households with larger families risk debt without income growth; education improvements could shift Under-Graduates to higher income brackets.
- Larger families (5–7 members) tend to have higher expenses (20,000–40,000) regardless of income, while smaller families (1–3 members) are more common and have more manageable expenses.

#### Weighted monthly sales by region

```python
filterSales = prices[prices['Attribute'] == 'Total weighted monthly sales in U.S. dollars']
plt.figure(figsize=(12, 6))
sns.barplot(data=filterSales, x='Metroregion_code', y='Value', palette='Set1')
plt.xticks(rotation=90)
plt.title('Total Weighted Monthly Sales by Metroregion')
plt.xlabel('Metroregion Code')
plt.ylabel('Total Weighted Monthly Sales (in U.S. dollars)')
plt.yscale('log')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

### Weighted Sales By Metroregion Analysis
- The total weighted monthly sales are scaled logarithmically, thus a jump from 10^7 to 10^8 is a 10x increase in sales.
- There is significant variation in sales across different metroregions, with some regions showing a 10x increase in sales compared to others.
- Factors such as population density, income levels, and food preferences may contribute to these differences.

```python
diet_hunger_group = habits.groupby(['What would best describe your diet:', 'Choose all that apply: [I experience feelings of hunger during the day]']).size().unstack(fill_value=0)
diet_hunger_group.plot(kind='bar', stacked=False, figsize=(10, 6),rot=90)
plt.title('Hunger Frequency Distribution by Diet Type')
plt.xlabel('Diet Type')
plt.ylabel('Count')
plt.legend(title='Hunger Frequency')
plt.xticks(rotation=90)
plt.show()
```

- The data shows that people on a vegetarian diet feel hungry more often than those on a non-vegetarian diet
    - the proportion of Eggetarians and Vegetarians who feel hungry very frequently are much higher than the rest, in fact they are the only two categories who reported feeling hungry very frequently
    - this may be due to the fact that vegetarian diets are often lower in protein and fat, which can lead to increased feelings of hunger and less satiety.
    - carbohydrate based diets also seem to influence ghrelin production and insulin resistance, which may also account for this increased hunger, or higher incidence of hunger pangs throughout the day
- pescatarians, or vegeterians who primarily eat seafood have the least incidence of hunger pangs, which may be due to the fact that seafood is often higher in protein and fat, leading to increased satiety and less frequent feelings of hunger

### Insights
- Foods higher in Fats and Proteins seem to have better outcomes in relation to hunger and frequency of feeding
- Foods higher in Carbohydrates seem to have worse outcomes in relation to hunger and frequency of feeding as well as potential insulin and ghrelin desensitivity
- This may be due to the fact that carbohydrate based diets are often lower in protein and fat, which can lead to increased feelings of hunger and less satiety.
